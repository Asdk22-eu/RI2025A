```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
file_path = "/01_corpus_turismo_500.txt"

# Leer el corpus
with open(file_path, "r", encoding="utf-8") as f:
    documentos = [line.strip() for line in f if line.strip() != ""]

print(f"Se han cargado {len(documentos)} documentos.")
```

```python
# Crear vectorizador
vectorizador = TfidfVectorizer()
# Ajustar al corpus y transformar
tfidf_matrix = vectorizador.fit_transform(documentos)
# Convertir a DataFrame para visualización
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizador.get_feature_names_out())
print(tfidf_df.head())
```
